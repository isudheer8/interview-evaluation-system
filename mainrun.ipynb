```python
import os
import sys
from pathlib import Path

def setup_project_paths():
    # 1. Start from current directory
    start_dir = Path(os.getcwd()).resolve()
    print(f"üîé Searching from: {start_dir}")
    
    # 2. Look for the 'core' folder by climbing up the tree
    current = start_dir
    found_root = None
    
    # Check 5 levels up max
    for _ in range(5):
        # Is 'core' inside the current folder?
        if (current / "core").is_dir():
            found_root = current
            break
        # Is 'core' inside a subfolder (like if you have a nested 'src' or 'app'?)
        elif (current / "interview-evaluation-system" / "core").is_dir():
            found_root = current / "interview-evaluation-system"
            break
        current = current.parent

    if found_root:
        root_str = str(found_root)
        if root_str not in sys.path:
            sys.path.insert(0, root_str) # Insert at 0 to prioritize this path
        
        print(f"‚úÖ FOUND ROOT: {root_str}")
        print(f"üìÇ CORE LOCATION: {found_root / 'core'}")
        
        # List contents to verify
        print(f"üìÑ Files in Core: {os.listdir(found_root / 'core')}")
    else:
        print("‚ùå CRITICAL: Could not find 'core' folder. Please check your folder names.")

setup_project_paths()

# Now test the import
try:
    from core.models.audio.dsp_stub import BasicDSP
    print("üöÄ Import Successful!")
except ImportError as e:
    print(f"üõë Still failing: {e}")
```
